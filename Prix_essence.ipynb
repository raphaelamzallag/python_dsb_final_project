```python
import pandas as pd
```

```python
df = pd.read_csv('prix_essence_finaaal.csv')

# Display information about the DataFrame
df
```

```python
new_df = df['date;prix'].str.split(';', expand=True)

# Rename the columns
new_df.columns = ['Date', 'Prix']

# Concatenate the new DataFrame with the original DataFrame
df_1 = pd.concat([df, new_df], axis=1)

# Drop the original 'Date;Prix' column if needed
df_2 = df_1.drop('date;prix', axis=1)

# Display the updated DataFrame
df_2
```
