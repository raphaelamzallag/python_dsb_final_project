```python
def prix_essence (X):
    df_essence['date'] = pd.to_datetime(df_essence['date'])
    pe = df_essence.rename(columns={'date': 'date_merge'})
    X['date_merge'] = pd.to_datetime(X['date'].dt.strftime('%Y-%m-%d'), format='%Y-%m-%d')
    X_pe = X.merge(pe, how = 'left', on = 'date_merge')
    X_pe = X_pe.drop(['date_merge'], axis=1)
    X_pe = X_pe.rename(columns={'date': 'date_merge'})
    X_pe = df_essence.rename(columns={'date': 'date_merge'})
    return X_pe
    
```
